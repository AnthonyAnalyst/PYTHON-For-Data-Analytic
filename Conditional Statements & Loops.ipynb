# Control Structures in Python

## A. Conditional Statements

```python
#1. Write a program that checks a student’s score and prints:

score = 0

if score >= 50:
    print("pass")
    
else:
    print("fail")
    
```

```python
#2. Write a program that checks a person’s age and prints:

age = 34

if age >= 18:
    print("Adult")
    
else:
    print("Minor")
```

```python
#3. Write a program that checks a number and prints:

number = -8

if number > 0:
    print("Positive")
    
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

```python
#4. Write a program that checks a student’s marks and prints:
#“Excellent” if marks are 80 and above
#“Good” if marks are between 60 and 79 
#“Needs Improvement” if marks are below 60

marks = 59

if marks >= 80:
    print("Excellent")
elif marks >= 60:
    print("Good")
else:
    print("Needs Improvement")

```

# B. LOOPS

```python
#5. Write a program that prints all numbers from 1 to 10.

for number in range(1, 11):
    print(number)
```

```python
#6. Write a program that prints all even numbers between 1 and 20.

for number in range(1, 21):
    
    if number % 2 == 0:
        print(number, "is an Even Number")
    else:
        print(number, "is an Odd Number")
```

```python
#7. Write a program that counts how many numbers between 1 and 50 are greater than 25, then prints the final count.

count = 0

for number in range(1, 51):
    if number > 25:
        count += 1
        
print("Total numbers greater than 25:", count)
```

```python
#8. Write a program that calculates and prints the sum of all numbers from 1 to 100.

total = 0

for number in range(1, 101):
    total += number
    
print("sum of numbers form 1 to 100:", total)

```
