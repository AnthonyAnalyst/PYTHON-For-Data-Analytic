```python
#Household Size Classification for Entire Community

household_sizes = [4, 7, 3, 9, 5, 6, 2]

for x in household_sizes:
    if x >= 6:
        print(x,"-Large household")
    else:
        print(x,"-Standard household")
```

```python
#Child Vaccination Status Check
child_ages = [2, 6, 1, 4, 8, 3, 5]

for x in child_ages:
    if x > 5:
        print("Age", x, "- Schedule vaccination")
    else:
        print("Age", x, "- Already vaccinated")
```

```python
#School Attendance Analysis

attendance_percentages = [92, 78, 85, 95, 88, 72, 90]

for attendance in attendance_percentages:
    if attendance >= 90:
        print(attendance, "% - Excellent")
        
    elif attendance >= 80:
        print(attendance, "% - Good")
        
    else:
        print(attendance, "% - Needs follow-up")
        
    
```

```python
#Income Survey Analysis

monthly_incomes = [4500, 12000, 7500, 3000, 15000, 8500]

for income in monthly_incomes:
    if income < 5000:
        print(income, "KES - Low income")
        
    elif income < 15000:
        print(income, "KES - Middle income")
        
    else:
        print(income, "KES - High income")
```

```python
#Food Security Assessment

meals_per_day = [3, 2, 1, 3, 2, 2, 1]

for meals in meals_per_day:
    if meals >= 3:
        print(meals, "meals - Food secure")
    elif meals == 2:
        print(meals, "meals - Moderately food insecure")
    else:
        print(meals, "meals - Severely food insecure")
```
